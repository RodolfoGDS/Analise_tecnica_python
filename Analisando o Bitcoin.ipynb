# Deocmposição da Série Temporal do BTC.


<h2> A decomposição é um procedimento matemático de dividir uma série temporal única em várias séries temporais diferentes.<h2>
<h2>Geralmente, a série temporal é dividida em 3 componentes: tendência, sazonalidade e flutuação aleatória(ruido).<h2>

<h2>Existe a decomposição aditiva e a decomposição multiplicativa.<h2>

<h3> O Modelo Aditivo representa séries temporais como adições de todos os três componentes:<h3>
Série Temporal = Tendência + Sazonal + Aleatória

<h3>O modelo multiplicativo representa as séries temporais como multiplicações de todos os três componentes:<h3>
Série Temporal = Tendência * Sazonal * Aleatória

```python
#Decomposição da Série Temporal do Bitcoin.

from statsmodels.tsa.seasonal import seasonal_decompose

btc[['Adj Close']].info()


# Modelo Aditivo
decomposicao_btc_aditiva = seasonal_decompose(btc[['Adj Close']], model= 'additive', period=100, extrapolate_trend=100)

# Modelo Multiplicativo
decomposicao_btc_multip = seasonal_decompose(btc[['Adj Close']], model = 'multiplicative', period= 100, extrapolate_trend=100)
```

```python
# Calculando a serie temporial a partir do modelo aditivo


serie_temporal_btc_aditivo = decomposicao_btc_aditiva.trend + decomposicao_btc_aditiva.seasonal + decomposicao_btc_aditiva.resid



# Calculando a Série temporal a partir do modelo 
serie_temporal_btc_multip = decomposicao_btc_multip.trend * decomposicao_btc_multip.seasonal * decomposicao_btc_multip.resid

display(serie_temporal_btc_aditivo, serie_temporal_btc_multip )
```

```python
#Plotando gráfico de tendencia, sazonalidade e e ruido da série temporal do BTC.


# Determinando o tamanho da figira e sua cor de fundo 
plt.figure(figsize=(16,14))
sns.set(style='darkgrid')


#Trend BTC
plt.subplot(3,1,1)
ax= sns.lineplot(data= decomposicao_btc_aditiva.trend)
plt.legend(labels=['Bitcoin - Tendencia'], loc=2, fontsize='large')
ax=ax


#Sazonalidade BTC
plt.subplot(3,1,2)
ax = sns.lineplot(data=decomposicao_btc_aditiva.seasonal)
plt.legend(labels=['Bitcoin - Sazonalidade'], loc=2, fontsize='large')
ax=ax

#Ruido BTC
plt.subplot(3,1,3)
ax= sns.lineplot(data = decomposicao_btc_aditiva.resid)
plt.legend(labels=['Bitcoin - Ruido'], loc = 2, fontsize='large')
ax=ax

plt.tight_layout()
```

```python
# Comparadndo o observado com a tendecia

ax, fig = plt.subplots(figsize=(15,8))
plt.plot(decomposicao_btc_aditiva.observed, color='blue', label='Observado')
plt.plot(decomposicao_btc_aditiva.trend, color= 'red', label= 'Tendencia')
ax.legend(labels= ['Observado', 'Tendencia']) 
ax=ax
```
