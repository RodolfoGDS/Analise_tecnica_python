# Análise de **ativos financeiros** com pyhton

### 1- Bitcoin - Dolar
### 2-Dolar - Real
### 3-Ouro
### 4-S&P500

```python
# Carregando as bibliotecas
from pandas_datareader import data as web
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# coletando dados de série historica dos 4 ativos citados
### ticker ---> BTC-USD .....ouro (GC=F)


btc = web.DataReader('BTC-USD', data_source='yahoo',start='17-09-2014', end='01-01-2022')
dolar = web.DataReader('BRL=X', data_source='yahoo', start='17-09-2014', end='01-01-2022')
ouro = web.DataReader('GC=F', data_source='yahoo', start='17-09-2014', end='01-01-2022')
sp_500 = web.DataReader('^GSPC',  data_source='yahoo', start='17-09-2014', end='01-01-2022')

```

```python
display(btc,dolar,ouro,sp_500)
```

```python
# Analisando serie historica dos ativos
plt.figure(figsize=(16,14))
sns.set(style='darkgrid')

#BTC
plt.subplot(4,1,1)
ax= sns.lineplot(data=btc, x=btc.index, y=btc['Adj Close'], color='orange')
plt.legend(labels=['Bitcoin'], loc=2, fontsize='large')
ax=ax

#Dolar
plt.subplot(4,1,2)
ax=sns.lineplot(data=dolar, x=dolar.index, y=dolar['Adj Close'], color='green')
plt.legend(labels=['Dolar'], loc=2, fontsize='large')
ax=ax

#Ouro
plt.subplot(4,1,3)
ax=sns.lineplot(data=ouro, x=ouro.index, y=ouro['Adj Close'], color= 'yellow')

plt.legend(labels=['Ouro'], loc=2, fontsize='large')
ax=ax

#SP&500
plt.subplot(4,1,4)
ax=sns.lineplot(data=sp_500,x=sp_500.index, y=sp_500['Adj Close'],color='black')
plt.legend(labels=['SP&500'], loc=2, fontsize='large')
ax=ax

plt.tight_layout()
```
